```python
#train-test splitting and scaling
```

```python
X = df.drop('species',axis=1)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=100)
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#finding best algorithm
```

```python
#logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
```

```python
LR.fit(X_train,y_train)
```

```python
y_pred=LR.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,plot_confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
plot_confusion_matrix(LR,X_test,y_test);
```

```python
#KNN
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier()
```

```python
KNN.fit(X_train,y_train)
```

```python
y_pred=KNN.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
plot_confusion_matrix(KNN,X_test,y_test);
```

```python
#SVM
```

```python
from sklearn.svm import SVC
svc=SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred=svc.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
plot_confusion_matrix(svc,X_test,y_test);
```

```python
#random forest
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred=rf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
plot_confusion_matrix(rf,X_test,y_test);
```

```python
#AdaBoost
```

```python
from sklearn.ensemble import AdaBoostClassifier
ab=AdaBoostClassifier()
```

```python
ab.fit(X_train,y_train)
```

```python
y_pred=ab.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
plot_confusion_matrix(ab,X_test,y_test);
```

```python
#SVM performed the best
#Hyperparameter tuning using Grid Search 
```

```python
param_grid= {'C':[0.1,0.5,1,1.5,2,5,10], 'kernel':['rbf','sigmoid','poly'], 'degree':[1,2,3], 'gamma':['scale','auto']}
```

```python
from sklearn.model_selection import GridSearchCV
grid_scv_model=GridSearchCV(estimator=svc,param_grid=param_grid,scoring='accuracy',cv=5)
```

```python
grid_scv_model.fit(X_train,y_train)
```

```python
grid_scv_model.best_estimator_
```

```python
y_pred=grid_scv_model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
plot_confusion_matrix(grid_scv_model,X_test,y_test);
```

```python
#final model
```

```python
final_model=SVC(C=1,gamma='auto')
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
final_model.fit(X_scaled,y)
```
