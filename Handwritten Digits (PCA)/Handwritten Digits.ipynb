```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
digits = pd.read_csv('digits.csv')
```

```python
pixels = digits.drop('number_label',axis=1)
```

### Scaling Data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaled_pixels = scaler.fit_transform(pixels)
```

```python
scaled_pixels
```

## PCA

```python
from sklearn.decomposition import PCA
```

```python
pca_model = PCA(n_components=2)
```

```python
pca_pixels = pca_model.fit_transform(scaled_pixels)
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

#### scatterplot of the digits in the 2 dimensional PCA space

```python
plt.figure(figsize=(10,6),dpi=150)
labels = digits['number_label'].values
sns.scatterplot(pca_pixels[:,0],pca_pixels[:,1],hue=labels,palette='Set1');
plt.legend(loc=(1.05,0));
```

## 4 and 6 are the most differentiated!

```python
# 3 Components

pca_model = PCA(n_components=3)
```

```python
pca_pixels = pca_model.fit_transform(scaled_pixels)
```

```python
from mpl_toolkits import mplot3d
```

```python
plt.figure(figsize=(8,8),dpi=150)
ax = plt.axes(projection='3d')
ax.scatter3D(pca_pixels[:,0],pca_pixels[:,1],pca_pixels[:,2],c=labels);
```
